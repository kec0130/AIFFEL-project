### 4.2.2. 이름을 띄어쓰기 & 대문자 기준으로 분리해 토큰화하기

```python
import re

name = "CharizardMega Charizard X"
name_split = name.split(" ")  # 이름을 공백 기준으로 쪼개기
name_split
```

```python
temp = name_split[0]  # 'CharizardMega'
tokens = re.findall('[A-Z][a-z]*', temp)  # 대문자로 시작해서 소문자로 끝나는 패턴 찾기
tokens
```

```python
# 토큰화 함수
def tokenize(name):
    name_split = name.split(" ")
    
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
    
    return np.array(tokens)
```

```python
name = "CharizardMega Charizard X"
tokenize(name)
```

```python
# 전설의 포켓몬 이름 토큰 모으기
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter

# 많이 사용된 토큰 Top 10 찾기
most_common = Counter(token_set).most_common(10)
most_common
```

```python
# 이름에 해당 토큰이 포함되는지 여부를 나타내는 컬럼 만들기
for token, _ in most_common:
    pokemon[token] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

# 5. 분류 모델 학습시키기
## 5.1. 원본 데이터로 베이스라인 모델 만들기

```python
print(original_data.shape)
original_data.head()
```

```python
original_data.columns
```

```python
# 의미 없는 컬럼과 문자열 feature 제외
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
```

```python
target = 'Legendary'
target
```

```python
X = original_data[features]
print(X.shape)
X.head()
```

```python
y = original_data[target]
print(y.shape)
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## 5.2. 의사 결정 트리 모델 학습시키기

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=25)
model
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

- 포켓몬 데이터는 불균형 데이터이므로 800마리를 전부 다 일반 포켓몬으로 찍어도 735 / 800 * 100 = 92% 의 정확도가 나온다. 따라서 정확도 외에 다른 척도로 모델의 성능을 평가해 볼 필요가 있다.
- 불균형 데이터에서는 적은 양의 데이터인 Positive(전설의 포켓몬)를 잘 잡아내는 것이 중요하다. 그런데 Recall이 낮다는 것은 분모에 있는 FN 값이 높다는 것을 의미한다. 즉, 전설의 포켓몬인데 일반 포켓몬이라고 판단하는 경우가 많은 것이다.
